```python
import re
def clean_text(text):
    text = text.lower()
    text = re.sub(r"[^a-z\s]", "", text)  
    text = re.sub(r"\s+", " ", text).strip()
    return text
df["cleaned_lyrics"] = df["lyrics"].apply(clean_text)
df.head()
```

```python
nltk.download("punkt")
nltk.download("punkt_tab")


df["tokens"] = df["cleaned_lyrics"].apply(word_tokenize)
display(df.head())
```

```python
from nltk.corpus import stopwords
nltk.download("stopwords")
stop_words = set(stopwords.words("english"))
df["tokens_no_stop"] = df["tokens"].apply(lambda x: [word for word in x if word not in stop_words])
```

```python
df.to_csv("cleaned_lyrics.csv", index=False)
```

```python
fig, axes = plt.subplots(1, 3, figsize=(15, 5))
for i, row in df.iterrows():
    freq = Counter(row["tokens_no_stop"]).most_common(15)
    words, counts = zip(*freq)
    axes[i].bar(words, counts)
    axes[i].set_title(f"{row['song']} Top Words")
    axes[i].tick_params(axis='x', rotation=45)
plt.tight_layout()
plt.show()
```

```python
from nltk.stem import WordNetLemmatizer
import seaborn as sns
def lemmatize_tokens(tokens):
    lemmatizer = WordNetLemmatizer()
    return [lemmatizer.lemmatize(word) for word in tokens]

def get_sentiment(text):
    return TextBlob(text).sentiment.polarity

def count_agency_words(tokens):
    agency_keywords = ["plan", "wishing", "make", "send", "right", "know", "feel", "love", "pain", "change"]
    return sum(1 for word in tokens if word.lower() in agency_keywords)
    
plt.figure(figsize=(8, 6))
colors = sns.color_palette("tab10", 3)
for i, row in df.iterrows():
    plt.scatter(row["sentiment"], row["agency_count"], s=100, c=[colors[i]], label=row["song"])
plt.xlabel("Sentiment Polarity (TextBlob)")
plt.ylabel("Agency Keyword Count")
plt.title("Sentiment vs. Agency in Hip-Hop Lyrics")
plt.legend()
plt.grid(True)
plt.show()
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
import seaborn as sns
import matplotlib.pyplot as plt


vectorizer = TfidfVectorizer(stop_words="english")
tfidf_matrix = vectorizer.fit_transform(df["cleaned_lyrics"])

cos_sim = cosine_similarity(tfidf_matrix, tfidf_matrix)


plt.figure(figsize=(6,4))
sns.heatmap(cos_sim, xticklabels=df["song"], yticklabels=df["song"], annot=True, cmap="YlOrRd")
plt.title("Similarity Between Songs")
plt.show()
```

```python
import networkx as nx
from nltk import bigrams
from collections import Counter
import matplotlib.pyplot as plt

tokens = df["tokens_no_stop"].sum() 
bigram_list = list(bigrams(tokens))
bigram_counts = Counter(bigram_list).most_common(20)
G = nx.Graph()
for (w1, w2), count in bigram_counts:
    G.add_edge(w1, w2, weight=count)
plt.figure(figsize=(8,6))
pos = nx.spring_layout(G, k=0.5)
nx.draw(G, pos, with_labels=True, node_size=700, node_color="teal", font_size=10, width=[d['weight'] for (_,_,d) in G.edges(data=True)])
plt.title("Word Connection Graph (Top 20 Bigrams)")
plt.show()
```

```python
themes = {
    "success": ["win", "success", "shine", "best", "king", "fame", "great"],
    "money": ["money", "cash", "dollar", "rich", "bank", "wealth", "ballin"],
    "struggle": ["pain", "hard", "struggle", "lose", "suffer", "tears", "sorry"],
    "love": ["love", "heart", "baby", "girl", "feel", "miss", "together"],
    "street_life": ["street", "hood", "hustle", "trap", "ride", "block", "corner"],
    "identity": ["self", "mirror", "know", "life", "real", "truth", "freedom"]
}
```

```python
def count_themes(tokens, themes):
    counts = {}
    for theme, keywords in themes.items():
        counts[theme] = sum(1 for word in tokens if word in keywords)
    return counts
theme_data = []
for i, row in df.iterrows():
    counts = count_themes(row["tokens_no_stop"], themes)
    counts["song"] = row["song"]
    counts["artist"] = row["artist"]
    theme_data.append(counts)
themes_df = pd.DataFrame(theme_data)
```

```python
themes_df.set_index("song")[list(themes.keys())].plot(kind="bar", figsize=(12,6))
plt.title("Recurring Themes Across Songs")
plt.ylabel("Word Count")
plt.xticks(rotation=45)
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import pandas as pd


if not isinstance(themes_df, pd.DataFrame) or "song" not in themes_df.columns or not themes:
    raise NameError("Ensure 'themes_df' is a DataFrame with a 'song' column and 'themes' is a dictionary with theme keywords.")


data = themes_df.set_index("song")[themes.keys()]
songs = data.index
themes_list = list(themes.keys())
n_songs = len(songs)
n_themes = len(themes_list)


fig = plt.figure(figsize=(8, 8))
ax = fig.add_subplot(111, projection='polar')


theta = np.linspace(0, 2 * np.pi, n_songs, endpoint=False)


width = (2 * np.pi) / n_songs * 0.8


colors = sns.color_palette("Blues", n_themes)


for i, theme in enumerate(themes_list):
    values = data[theme].values
    ax.bar(theta + i * (width / n_themes), values, width=width / n_themes, label=theme, color=colors[i], alpha=0.7)


ax.set_xticks(theta)
ax.set_xticklabels(songs, fontsize=10)
ax.set_title("Theme Distribution in Songs (Radial)", pad=20)
ax.legend(loc='upper right', bbox_to_anchor=(1.1, 1.1))
plt.show()
```
